```python
# STEP 1: Install OpenCV (if not already)
!pip install opencv-python

# STEP 2: Upload an image
from google.colab import files
uploaded = files.upload()

import cv2
import numpy as np
from matplotlib import pyplot as plt
from PIL import Image
import io

# STEP 3: Read and convert the uploaded image
filename = next(iter(uploaded))  # Get the uploaded filename
image_stream = io.BytesIO(uploaded[filename])
img = Image.open(image_stream).convert('RGB')
img_np = np.array(img)

# STEP 4: Convert to BGR for OpenCV
img_cv = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)
gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)

# STEP 5: Load Haar Cascade and detect faces
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, 1.1, 5)

# STEP 6: Draw rectangles around faces
for (x, y, w, h) in faces:
    cv2.rectangle(img_cv, (x, y), (x+w, y+h), (255, 0, 0), 2)

# STEP 7: Show the result
plt.figure(figsize=(8,6))
plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
plt.title("Detected Faces")
plt.axis('off')
plt.show()
```
